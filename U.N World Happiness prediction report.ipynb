## Constructing three models without tuning
    I started by building three simple models: random forest classifier, GradientBoosting classifier and Keras deep learning model. ## Tuning the models 
    I started with random forest classifier model. I used random grid with 3 fold cross validation to search for the best hyperparameters. However, it did not improve the RFC model performance. The tuned RFC model still only recivew accuracy 0.5. 
    Then I tried to tune the Keras deep learning model. Since this is my first time using deep learning method to build a predictive model, I was able to find some best hyperparameters like 'batch_size = 64' and 'epochs = 10', but the improvement was limited and still worse than RFC. 

## Best model
    The best model was the tuned GradientBoosting classifier. I used gridseach cv to find the hyperparameters of learning_rate': 0.1, 'max_depth': 5, 'n_estimators': 50, 'subsample': 0.5. The accuracy score of the train set was successfully improved to 0.656481. 
    However, when I submited the tuned GB model, it scored exactly the same as the untuned GB model. I tried to figure out the reason because there is clearly improvement after tuning the hyperparameters but I failed. I also tried to rewrite the code few times but did not find the answer. 

## Conclusion
    In this assignment, the train dataset is very small, so Keras deep learning method performed bad. The target variable "happiness" is categorical instead of numerical value, I guess that is the reason why tuning the GB model seemed no improvement. In other words, the improvement is not large enough to change the predictive value, for example from 'low' to 'average'.
